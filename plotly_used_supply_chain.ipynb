```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/content/supply_chain - supply_chain.csv")
df
```

```python
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
pio.templates.default = "plotly_dark"
```

```python
df.describe()
```

```python
fig=px.scatter(df,x='Price',
               y='Revenue generated',
               color='Product type',
               hover_data=['Number of products sold'],
               trendline='ols')
fig.show()
```

```python
df.columns
```

```python
sales_data=df.groupby('Product type')['Number of products sold'].sum().reset_index()
```

```python
pie_chart=px.pie(sales_data,values='Number of products sold',
                 names='Product type',
                 title='Sales by Product Type',
                 hover_data=['Number of products sold'],
                 hole=0.5,
                 color_discrete_sequence=px.colors.qualitative.Pastel)
pie_chart.update_traces(textposition='inside',textinfo='percent+label')
pie_chart.show()
```

```python
total_revenue=df.groupby('Shipping carriers')['Revenue generated'].sum().reset_index()
```

```python
fig=go.Figure()
fig.add_trace(go.Bar(x=total_revenue['Shipping carriers'],
                     y=total_revenue['Revenue generated'])),
fig.update_layout(title='Total Revenue by Shipping Carrier',
                                       xaxis_title='Shipping Carrier',
                                       yaxis_title='Revenue Generated')
fig.show()
```

```python
#so the company is using three carriers for transportation,and carrier B helps the company in generationg more revenue.now its have a
#look at the Average lead time and average manufacturing costs for all products of the company
```

```python
avg_lead_time=df.groupby('Product type')['Lead time'].mean().reset_index()
```

```python
avg_manu_costs=df.groupby('Product type')['Manufacturing costs'].mean().reset_index()
```

```python
a=pd.merge(avg_lead_time,avg_manu_costs,on='Product type')
```

```python
a=a.rename(columns={'Lead time': 'Avg Lead time','Manufacturing costs':'Avg Manu costs'})
a
```

```python
fig=go.Figure()
fig.add_trace(go.Bar(x=a['Product type'],
                     y=a['Avg Lead time'])),
fig.update_layout(title='Average Lead Time by Product Type',
                                       xaxis_title='Product type',
                                       yaxis_title='Avg Lead time')
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Bar(x=a['Product type'],
                     y=a['Avg Manu costs'])),
fig.update_layout(title='Average Manu cost by Product Type',
                                       xaxis_title='Product type',
                                       yaxis_title='Avg Manu costs')
fig.show()
```
