```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/tips - tips.csv")
df.head(4)
```

```python
x=df.drop(columns=['time'])
x.head(3)
y=df['time']
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Male','Female'],['Yes','No']])
x_train_sex_smoker=oe.fit_transform(x_train[['sex','smoker']])
x_train_sex_smoker.shape



```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop='first',sparse_output=False)
x_train_day=ohe.fit_transform(x_train[['day']])
x_train_day.shape
```

```python
x_train_transformed=np.concatenate((x_train_day,x_train_sex_smoker),axis=1)
```

```python
x_train_transformed.shape
```

```python
from sklearn.compose import ColumnTransformer
Transformer=ColumnTransformer(transformers=[
      #('tnf1',SimpleImputer(),['fever']),
     ('tnf2',OrdinalEncoder(categories=[['Male','Female'],['Yes','No']]),['sex','smoker']),
     ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['day'])
     ],remainder='passthrough')

```
