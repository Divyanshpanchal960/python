```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df=pd.read_csv("/content/Attrition - Attrition.csv")
df
```

```python
df.shape
df.columns
```

```python
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default="plotly_white"
```

```python
#Filter the data to show only 'Yes' values in the 'Attrition' column
df1=df[df['Attrition']=='Yes']
df1
```

```python
#department wise attrition
df1['Department'].value_counts()
```

```python
df1.groupby('Department').size().reset_index(name='count')  #attrition employees
```

```python
#attrition by department
#create a donut chart
attrition_counts = df1.groupby('Department').size().reset_index(name='count')
fig=go.Figure(data=[go.Pie(
    labels=attrition_counts['Department'],
    values=attrition_counts['count'],
    hole=0.6,
    textposition='inside'
)])
fig.update_layout(title_text='Attrition by Department')
fig.show()
```
