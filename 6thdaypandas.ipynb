```python
import pandas as pd

data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],}

data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']}

df1 = pd.DataFrame(data1)

df2 = pd.DataFrame(data2)

print(df1)
print(df2)
```

```python
#perform merging based on key
rs=pd.merge(df1,df2,on='key')
rs
```

```python
#merging with different keys
import pandas as pd

data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K1', 'K0', 'K1'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],}

data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K0', 'K0', 'K0'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']}

df1 = pd.DataFrame(data1)

df2 = pd.DataFrame(data2)

print(df1)
print(df2)
```

```python
rs=pd.merge(df1,df2,on=['key1','key'])
rs
```

```python
rs=pd.merge(df1,df2,how='left',on=['key1','key'])
rs
```

```python
rs=pd.merge(df1,df2,how='right',on=['key1','key'])
rs
```

```python
#Union
rs=pd.merge(df1,df2,how='outer',on=['key1','key'])
rs
```

```python
#intersection
rs=pd.merge(df1,df2,how='inner',on=['key1','key'])
rs
```
