```python
#Label_Encoding------> label encoding is a data processing technique used in machine learning to convert categorical data into a numerical formate
import numpy as np
import pandas as pd
df=pd.read_csv("/content/covid_toy - covid_toy.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.head(3)
```

```python
x=df.drop(columns=['has_covid'])
x.head(2)
y=df['has_covid']
y.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
df.head(2)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=23)
```
