```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/covid_toy - covid_toy.csv")
df.head(2)
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer(strategy='mean')
df['fever']=si.fit_transform(df[['fever']])
df.isnull().sum()
```

```python
df.head(3)
```

```python
df['city'].value_counts()
```

```python
df['gender']=df['gender'].map({'Male':1,'Female':2})
df['cough']=df['cough'].map({'Mild':1,'Strong':2})
df['city']=df['city'].map({'Kolkata':1,'Bangalore':2,'Delhi':3,'Mumbai':4})
df['has_covid']=df['has_covid'].map({'Yes':1,'No':2})
```

```python
np.round(df.describe(),2)
```

```python
x=df.drop(columns=['city'])
x.head(3)
y=df['city']
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=32)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_train_new=pd.DataFrame(x_train_sc,columns=x.columns)
np.round(x_train_new.describe(),1)
```

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
x_train_ms=ms.fit_transform(x_train)
x_train_new=pd.DataFrame(x_train_ms,columns=x.columns)
np.round(x_train_new.describe(),2)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x_train['age']=le.fit_transform(x_train['age'])
x_train['cough']=le.fit_transform(x_train['cough'])
x_train['fever']=le.fit_transform(x_train['fever'])
x_train['gender']=le.fit_transform(x_train['gender'])
x_train['has_covid']=le.fit_transform(x_train['has_covid'])
x_train.head(3)
```

```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/click - click.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Ad Topic Line']=le.fit_transform(df['Ad Topic Line'])
df['City']=le.fit_transform(df['City'])
df['Gender']=le.fit_transform(df['Gender'])
df['Country']=le.fit_transform(df['Country'])
df['Timestamp']=le.fit_transform(df['Timestamp'])
```

```python
df.head(3)
```

```python
x=df.drop(columns=['Clicked on Ad'])
x.head(4)
y=df['Clicked on Ad']
y.head(2)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=34)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_train_new=pd.DataFrame(x_train_sc,columns=x.columns)
np.round(x_train_new.describe(),2)
```

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
x_train_ms=ms.fit_transform(x_train)
x_train_new=pd.DataFrame(x_train_ms,columns=x.columns)
np.round(x_train_new.describe(),2)
```
