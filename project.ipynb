```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/content/archive.zip")
df
```

```python
df.head()
df.tail()
df['coffee_name'].value_counts()
#most selling drink
```

```python
df['coffee_name'].value_counts().plot(kind='bar',color=['red','green','blue','yellow','indigo','maroon','olive','black'])
# #sns.set(style='whitegrid')
# #plt.figure(figsize=(9,5))
# #sns.count(data=df,x= 'coffee_name')
plt.title("top selling drink")
plt.xlabel('coffee_name')
plt.ylabel('Count')
 #df.color=(['red','green','blue','yellow','indigo','maroon','olive','black'])
plt.show()
# #most selling drink chart
# #most selling drink(coffee_name=Americano with milk )
```

```python
print(df.groupby('coffee_name')['money'].sum().sort_values(ascending=False))
df.groupby('coffee_name')['money'].sum().sort_values(ascending=False).plot(kind='bar',color=(['red','green','blue','yellow','indigo','maroon','olive','black'])) #according to revenue/value most selling coffee of the coffee_shop, is latte
```

```python
#most selling drink percentage
df['coffee_name'].value_counts().plot(kind='pie',autopct='%.3f')
plt.title('most selling drink percantage')
plt.xlabel('coffee_name')
plt.ylabel('count')
plt.show()
#most selling drink parcentange chart according to quantity of selling
```

```python
print(df.groupby('hour_of_day')['money'].sum())
# print(df.groupby('hour_of_day')['money'].sum().plot(kind='bar'))
# print(df.groupby('hour_of_day')['money'].sum().plot(kind='barh'))
print(df.groupby('hour_of_day')['money'].sum().plot(kind='line'))
plt.title("peak business hour")
plt.xlabel("hour of day")
plt.ylabel("hour wise sum of money(money)")
plt.show()
#money according to hour wise -----> sabse jyda transaction 10 baje huaa hai (peak business hour=10oclock)
```

```python
print(df['Time_of_Day'].value_counts()) #peak purchasing time of day 'afternoon'
print(df.groupby('Time_of_Day')['money'].sum()) #peak transaction time of day is 'night'
print(df.groupby('Time_of_Day')['money'].sum().sort_values(ascending=True).plot(kind='bar',color=['tan','teal','pink'])) #peak selling by time of day
plt.title("Sales by time of day")
plt.xlabel("Time of day")
plt.ylabel("time wise sum of money")
plt.show()
```

```python
#df['Time_of_Day'].value_counts().plot(kind='pie',autopct='%.2f')
df.groupby('Time_of_Day')['money'].sum().plot(kind='pie',autopct='%.2f')
plt.title("percentage of sales by time of day")
plt.xlabel("time of day")

```

```python
df['Weekday'].value_counts()
df.groupby('Weekday')['money'].sum().sort_values(ascending=True).plot(kind='bar',color=['tan','teal','pink','grey','black','cyan','grey'])
plt.title("sales by weekday")
plt.xlabel("Weekday")
plt.ylabel("Weekday wise sum of money")
plt.show()
```

```python
print(df['Month_name'].value_counts()) #values count according to month
print(df.groupby('Month_name')['money'].sum())  #total sum of money according to month(highely revenue month)
df.groupby('Month_name')['money'].sum().sort_values(ascending=True).plot(kind='bar',color=['green','yellow','black','magenta','red','indigo','cyan','olive','maroon','teal','tan','orange']) #Top selling month of coffee is 'march'
plt.title("seasonal demand pattern")
plt.xlabel("sales by month")
plt.ylabel("money")
plt.show()
```
