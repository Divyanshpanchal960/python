```python
#Dataframe---------
import pandas as pd


a={  "Emp-id":[1,2,3,4,5],
      "name":["sam","raju","hari","ram","rohan"],
     "roles":["IT","sales","sales","IT","IT"],
     "working-hour":[8,8,8,9,9],
     "salary":["20000","20000","30000","40000","30000"]
}
```
